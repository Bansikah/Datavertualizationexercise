```python
fraudulent_transactions = 50
non_fraudulent_transactions = 950
```

```python
labels = ['Fraudulent', 'Non-Fraudulent']
sizes = [fraudulent_transactions, non_fraudulent_transactions]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # To highlight the 'Fraudulent' slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.title('Proportion of Fraudulent vs. Non-Fraudulent Transactions')
plt.show()
```

```python
transaction_types = ['Type A', 'Type B', 'Type C', 'Type D']
fraud_counts = [10, 20, 5, 15]

plt.bar(transaction_types, fraud_counts)
plt.xlabel('Transaction Type')
plt.ylabel('Number of Fraudulent Transactions')
plt.title('Number of Fraudulent Transactions by Transaction Type')
plt.grid(True)
plt.show()
```
